Harry Potter

```python
response2 = requests.get('https://wizard-world-api.herokuapp.com/Houses')
results2 = response2.json()
results2
```

```python
flattened_data = json_normalize(results2)

flattened_data['heads'][0]
```

```python
flattened_data['heads'][2]
```

```python
def extract_heads(x):
    heads = []
    for dictionary in x:
        heads.append(dictionary)
    print(heads)
    return heads
```

```python
extract_heads([{'id': '530da97d-5a83-4ea6-bc15-790edf5b5efc',
  'firstName': 'Minerva',
  'lastName': 'McGonagall'},
 {'id': '9915c5f8-9177-4f63-bba8-d04387a404f9',
  'firstName': 'Godric',
  'lastName': 'Gryffindor'}])
```

```python
flattened_data['House Heads'] = flattened_data['heads'].apply(str)
```

```python
flattened_data.head()
```

```python
import re

def extract_house_heads(x):
    text = " ".join([str(value) for value in x.values() if isinstance(value, str)])
    text = re.sub(r'[^a-zA-Z\s]', '', text)  
    return 'id: ' + text.strip()

flattened_data['Head House 1'] = flattened_data['heads'].apply(lambda x: extract_house_heads(x[0]) if len(x) > 0 else None)
flattened_data['Head House 2'] = flattened_data['heads'].apply(lambda x: extract_house_heads(x[1]) if len(x) > 1 else None)
```

```python
flattened_data.head()
```

```python
# flattened_data['House Head 1'] = flattened_data['heads'].apply(lambda x: str(x[0]) if len(x) > 0 else None)
# flattened_data['House Head 2'] = flattened_data['heads'].apply(lambda x: str(x[1]) if len(x) > 1 else None)
```

```python
flattened_data['House Trait 1'] = flattened_data['traits'].apply(lambda x: extract_house_heads(x[0]) if len(x) > 0 else None)
flattened_data['House Trait 2'] = flattened_data['traits'].apply(lambda x: extract_house_heads(x[1]) if len(x) > 0 else None)
flattened_data['House Trait 3'] = flattened_data['traits'].apply(lambda x: extract_house_heads(x[2]) if len(x) > 0 else None)
flattened_data['House Trait 4'] = flattened_data['traits'].apply(lambda x: extract_house_heads(x[3]) if len(x) > 0 else None)
flattened_data['House Trait 5'] = flattened_data['traits'].apply(lambda x: extract_house_heads(x[4]) if len(x) > 0 else None)
```

```python
flattened_data.head()
```

```python
flattened_data.drop(['heads', 'traits'], axis=1)
```

Public Holidays Germany

```python
response3 = requests.get('https://date.nager.at/api/v3/publicholidays/2023/DE')
results3 = response3.json()
results3
```

```python
public_holidays = json_normalize(results3)
```

```python
public_holidays.head()
```

```python
def extract_text_from_list(x):
    if x:
        return ' '.join(x)
    else:
        return 'no info'
```

```python
public_holidays['counties'] = public_holidays['counties'].apply(extract_text_from_list)
public_holidays['types'] = public_holidays['types'].apply(extract_text_from_list)
```
